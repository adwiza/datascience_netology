## Добавляем случайный шум на изображение

```python
noise_img = img.astype(float) + np.random.uniform(img.astype(float))
noise_img = np.uint8(255. * np.abs(noise_img) / np.max(np.abs(noise_img)))
plt.imshow(noise_img[...,::-1])
```

```python
from ipywidgets import interact

def median_filter_show(ksize=5):
    filtered = median_filter(noise_img, ksize)
    return plt.imshow(filtered[...,::-1]);
interact(median_filter_show, ksize=(1, 20, 1))
```
