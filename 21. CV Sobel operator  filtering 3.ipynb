## Оператор Собеля

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('img/cv/lena.png')
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(img_grey, cmap='gray');
```

## Оператор Собеля для вычисления X и Y составляющих градиента

```python
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])
```

## Вычислим градиенты по осям

```python
grad_x = cv2.filter2D(img_grey, cv2.CV_32F, sobel_x) # 8U
grad_y = cv2.filter2D(img_grey, cv2.CV_32F, sobel_y) # 8U
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(121)
plt.imshow(grad_x, cmap='gray')
plt.title('Grad X')
plt.subplot(122)
plt.imshow(grad_y, cmap='gray')
plt.title('Grad Y');
```

## Вычисляем суммарный градиент

```python
grad = np.sqrt(grad_x ** 2 + grad_y ** 2).astype(np.float32)
```

```python
plt.figure(2, figsize=(12, 8))
plt.subplot(121)
plt.imshow(img_grey, cmap='gray')
plt.subplot(122)
plt.imshow(grad, cmap='gray');
```

## cv2.Sobel

```python
grad_x = cv2.Sobel(img_grey, cv2.CV_32F, 2, 0, ksize=5)
grad_y = cv2.Sobel(img_grey, cv2.CV_32F, 0, 2, ksize=5)
```

```python
grad = np.sqrt(grad_x ** 2 + grad_y ** 2).astype(np.float32)
```

```python
plt.figure(2, figsize=(12, 8))
plt.subplot(121)
plt.imshow(img_grey, cmap='gray')
plt.subplot(122)
plt.imshow(grad, cmap='gray');
```
